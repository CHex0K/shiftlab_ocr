```python
import shift_ocr
import matplotlib.pyplot as plt
import numpy as np
```

```python
# download a model from here https://drive.google.com/file/d/1VFCOCSjw7021NrJcnSao5UKeIiNj_A8N/view?usp=sharing
PATH_TO_WEIGHTS = '/content/ocr_transformer_rn50_64x256_53str_jit.pt'
PATH_TO_IMAGE = '/content/test2.jpg'
scanner = shift_ocr.Scanner(PATH_TO_WEIGHTS)
result = scanner.doc2text(PATH_TO_IMAGE)
```

```python
result
```

```python
fig=plt.figure(figsize=(8, 8))
rows = int(len(result[1])/4) + 2
columns = int(len(result[1])/8) + 2
for j,crop in enumerate(result[1]):
  fig.add_subplot(rows, columns, j+1)
  plt.imshow(np.asarray(crop.img)) 
```
